#### 1. Загрузить файл wines_data.csv

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
wines=pd.read_csv('wine_data.csv')
```

```python
wines.describe()
```

#### 2. Рассчитать и визуализировать корреляции для переменных данного датасета

```python
corr = wines.corr()
corr
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

#### 3. Визуализировать взаимосвязи между 3-4 переменными

```python
sns.pairplot(wines, hue='Class')
```

```python
plt.scatter(wines['Alcohol'],wines['Total phenols'])
plt.xlabel('Alcohol')
plt.ylabel('Total phenols')
```

```python
plt.scatter(wines['Alcalinity of ash'],wines['Proline'], c="r",marker="*")
plt.xlabel('Alcalinity of ash')
plt.ylabel('Proline')
```

```python
plt.scatter(wines['Proanthocyanins'],wines['Flavanoids'], c="y",marker="2")
plt.xlabel('Proanthocyanins')
plt.ylabel('Flavanoids')
```

```python
#Проверим, например, что средняя крепость близка к 13
stats.ttest_1samp(wines['Alcohol'], 13)
```

```python
stats.ttest_ind(wines['Malic acid'], wines['Ash'])
```

```python
stats.ttest_ind(wines['Alcohol'], wines['Malic acid'])
```
