```python
def linreg(x,y):
    """
    Параметр None  у columns означается что все категорийные переменые он
    будет превращать
    в булевы """
   

    X = pd.get_dummies(x, columns = None) 
    
    from sklearn.linear_model import LogisticRegression
    model = LogisticRegression()
    model.fit( X, y )
    predictions = model.predict_proba( X )
    return model.score(X, y)
```

```python
#Переберем все комбинации параметров, вернем для низ оценку точности в словарь scores_train
import itertools
scores_train={}
stuff = all_columns
for L in range(1, len(stuff)+1):
    for subset in itertools.combinations(stuff, L):
        #print(data[list(subset)].head(1))
        vars_=subset
        scores_=linreg(X_train[list(subset)],y_train)
        scores_train.update({vars_:scores_})
        #print(len(scores_train))# просто смотрю что дело движется
```

```python
#Отсортируем по возрастанию наши оценки

scores_train_sort=sorted(scores_train.items(), key=lambda x:x[1], reverse=True)
```
