## Градиентный спуск

Vanilla update<br>
x += - learning_rate * dx

```python
#Рассчет логистической функции
def sigmoid (z):
    return 1 / (1 + np.exp(-z))
#Рассчет функции потерь
def loss(h,y):
    return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
# функция предсказания вероятности
def predict_proba(X,theta,fit_intercept=True):
    if len(theta)-1==X.shape[1]:
        intercept = np.ones((X.shape[0], 1))
        X=np.concatenate((intercept, X), axis=1)
    z = np.dot(X, theta)
    return sigmoid(z)
# функция предсказания класса
def predict(prediction_proba,rate=0.5):
    return np.asarray([0 if x<rate else 1 for x in prediction_proba])
```

```python
def log_reg_GD(X,y,fit_intercept=True,EPOCHS=500,LEARNING_RATE=0.01,EPS=0.000000003):
    # Если хотим добавить свободный член, добавляем к фичам массив единиц
    if fit_intercept==True:
        intercept = np.ones((X.shape[0], 1))
        X=np.concatenate((intercept, X), axis=1)
    theta = np.ones(X.shape[1])# будем считать начальной точкой массив единиц
    costs = []
    preds = []
    for _ in range(EPOCHS):
        
        prediction=predict_proba(X,theta) # Сформируем прдсказания
        gradient = np.dot(X.T, (prediction- y)) / y.size #Расчсчитаем градиент
        theta -= LEARNING_RATE * gradient # сдвинем оценки коэффцииентов
        
        #Сохраним разные промежутчоные предсказания и потери
        prediction=predict_proba(X,theta)
        preds.append(prediction)
        cost = loss(prediction, y)
        costs.append(cost)
        
        # Выйдем из цикла , если потери изменяются незначительно
        try:
            #print(abs(cost-costs[-2]))
            if abs(cost-costs[-2])<EPS:
                break
        except:
            pass
    return  theta, costs,  prediction  
```

```python
#заппустим на
model_gd=log_reg_GD(X_train,y_train,EPOCHS=1000000,LEARNING_RATE=0.001)
```

## Momentum

Momentum update<br>
v = mu * v - learning_rate * dx  # integrate velocity<br>
x += v # integrate position

```python
def log_reg_momentum(X,y,fit_intercept=True,EPOCHS=500,LEARNING_RATE=0.01,EPS=0.000000003,gamma=0.9):
    if fit_intercept==True:
        intercept = np.ones((X.shape[0], 1))
        X=np.concatenate((intercept, X), axis=1)
    theta = np.ones(X.shape[1])
    costs = []
    preds = []
    v=0
    vs=[]
    for _ in range(EPOCHS):
        prediction=predict_proba(X,theta)
        gradient = np.dot(X.T, (prediction- y)) / y.size
        v = gamma*v - LEARNING_RATE*gradient
        
        vs.append(v)
        theta += v
        
        prediction=predict_proba(X,theta)
        preds.append(prediction)
        cost = loss(prediction, y)
    
        costs.append(cost)
        try:
            
            if abs(cost-costs[-2])<EPS:
                break
        except:
            pass
    return  theta, costs,  prediction ,vs
```

## NAG

evaluate dx_ahead (the gradient at x_ahead instead of at x)<br>
v = mu * v - learning_rate * dx_ahead<br>
x += v

```python
def log_reg_NAG(X,y,fit_intercept=True,EPOCHS=500,LEARNING_RATE=0.01,EPS=0.000000003,gamma=0.9):
    if fit_intercept==True:
        intercept = np.ones((X.shape[0], 1))
        X=np.concatenate((intercept, X), axis=1)
    theta = np.ones(X.shape[1])
    costs = []
    preds = []
    v=0
    vs=[]
    for _ in range(EPOCHS):
        prediction=predict_proba(X,theta)
        gradient = np.dot(X.T, (prediction- y)) / y.size
        v_prev = v
        v = gamma*v - LEARNING_RATE*gradient
        #print(v)
        vs.append(v)
        theta += -gamma*v_prev+(1+gamma) *v
        
        prediction=predict_proba(X,theta)
        preds.append(prediction)
        cost = loss(prediction, y)
    
        costs.append(cost)
        try:
            
            if abs(cost-costs[-2])<EPS:
                break
        except:
            pass
    return  theta, costs,  prediction ,vs
```

## Adagrad

Adagrad
 Assume the gradient dx and parameter vector x<br>
cache += dx**2<br>
x += - learning_rate * dx / (np.sqrt(cache) + eps)

```python
def log_reg_Adagrad(X,y,fit_intercept=True,EPOCHS=500,LEARNING_RATE=0.01,EPS=0.000000003,gamma=0.9):
    if fit_intercept==True:
        intercept = np.ones((X.shape[0], 1))
        X=np.concatenate((intercept, X), axis=1)
    theta = np.ones(X.shape[1])
    costs = []
    preds = []
    
    cache=np.zeros(X.shape[1])
    caches=[]
    for _ in range(EPOCHS):
        prediction=predict_proba(X,theta)
        gradient = np.dot(X.T, (prediction- y)) / y.size
        
        cache += gradient**2
        theta += - LEARNING_RATE*gradient/(np.sqrt(cache) + EPS)
        
        prediction=predict_proba(X,theta)
        preds.append(prediction)
        cost = loss(prediction, y)
        #print('cost',cost)
        costs.append(cost)
        try:
            
            if abs(cost-costs[-2])<EPS:
                break
        except:
            pass
    return  theta, costs,  prediction ,caches
```

## RMSprop

cache = decay_rate * cache + (1 - decay_rate) * dx**2<br>
x += - learning_rate * dx / (np.sqrt(cache) + eps)

```python
def log_reg_RMSprop(X,y,fit_intercept=True,EPOCHS=500,LEARNING_RATE=0.01,EPS=0.000000003,decay_rate=0.9):
    if fit_intercept==True:
        intercept = np.ones((X.shape[0], 1))
        X=np.concatenate((intercept, X), axis=1)
    theta = np.ones(X.shape[1])
    costs = []
    preds = []
    
    cache=np.zeros(X.shape[1])
    caches=[]
    for _ in range(EPOCHS):
        prediction=predict_proba(X,theta)
        gradient = np.dot(X.T, (prediction- y)) / y.size
        
        cache = decay_rate*cache+ (1 - decay_rate) * gradient**2
        caches.append(cache)
        
       
        theta += - LEARNING_RATE*gradient/(np.sqrt(cache) + EPS)
       
        prediction=predict_proba(X,theta)
        preds.append(prediction)
        cost = loss(prediction, y)
        costs.append(cost)
       
        try:
            
            if abs(cost-costs[-2])<EPS:
                break
        except:
            pass
    return  theta, costs,  prediction ,caches
```
