```python
import numpy as np
```

```python
def makeMean(Dataframe,ser):
    Dataframe[ser].fillna(Dataframe[ser].mean(), inplace=True)
        

        
def makeMostOften(Dataframe,ser):
    Dataframe[ser].fillna(Dataframe[ser].mode()[0], inplace=True)
        
```

```python

for row in df_cl.columns:
    if df_cl[row].dtype=='float64':
        
        makeMean(df_cl,row)
       
        
    else:
        
        makeMostOften(df_cl,row)
        
```
